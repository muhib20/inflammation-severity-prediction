```python
from xgboost import XGBRegressor
```

```python
xgb_reg = XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    objective="reg:squarederror",
    random_state=42
)
```

```python
xgb_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("model", xgb_reg)
])
```

```python
gkf = GroupKFold(n_splits=5)

mae_scores = []
```

```python
for train_idx, test_idx in gkf.split(X, y, groups=pat_id):
    X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
    y_tr, y_te = y.iloc[train_idx], y.iloc[test_idx]

    xgb_pipe.fit(X_tr, y_tr)        # TRAIN
    preds = xgb_pipe.predict(X_te) # PREDICT severity

    mae = mean_absolute_error(y_te, preds)
    mae_scores.append(mae)
```

```python
print("XGBoost REGRESSION (Severity 0–5) with GroupKFold")
print(f"MAE: {np.mean(mae_scores):.3f} ± {np.std(mae_scores):.3f}")
```
